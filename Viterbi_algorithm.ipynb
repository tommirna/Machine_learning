```python
def viterbi_algorithm(init_prob, transition_matrix, emission_matrix, obs_sequence):
    
    # Initialise current probabilities for states 
    current_prob_states = init_prob 
    
    # Calculate first round probabilities 
    num_states = init_prob.shape[0]
    
    # First observed state 
    obs_first = obs_sequence[0]
    
    # Find maximal probability for 
    # observed state 
    for i in range(0, num_states):
        state_prob = np.zeros(num_states)
        for j in range(0, num_states):
            
            # Probability from previous states and emission 
            state_prob[i] = current_prob_states[j] * emission_matrix[i, obs_first]
            
        # Update the current probabilities with max 
        current_prob_states[i] = max(state_prob)
            
    # Proceed the next observed states 
    for i in range(1, num_states):
        # Obs state 
        obs_state = obs_sequence[i]
        
        state_prob = np.zeros(num_states)
        for j in range(0, num_states):
            
            # Probability from previous states and emission 
            state_prob[i] = transition_matrix[i,j] * emission_matrix[i, obs_state]
            
        # Update the current probabilities with max 
        current_prob_states[i] = max(state_prob)
        
    return(current_prob_states)
    
    
```
